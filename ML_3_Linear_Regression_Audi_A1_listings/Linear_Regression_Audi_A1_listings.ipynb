```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Audi_A1_listings.csv")
```

```python
df.head(3)
```

```python
df.drop(["index","href","MileageRank","PriceRank","PPYRank","Score"],axis=1,inplace=True)
df.head(3)
```

```python
df.rename(columns={"Price(Â£)":"Price"},inplace=True)
df.head(3)
```

```python
df["Engine"] = df["Engine"].str.replace("L","")
df.head(3)
```

```python
df.info()
```

```python
df["Engine"] = pd.to_numeric(df["Engine"])
df.head(3)
```

```python
df["Type"].nunique()
```

```python
df.drop("Type",axis=1,inplace=True)
df.head(3)
```

```python
pd.get_dummies(df, columns=["Transmission","Fuel"])
```

```python
 df = pd.get_dummies(df, columns=["Transmission","Fuel"],drop_first=True)
 df.head(3)
```

```python
y = df["Price"]
x = df.drop("Price",axis=1)
```

```python
model = LinearRegression().fit(x,y)
```

```python
model.score(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(14,5))
sns.regplot(x=y,y=y_pred)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("heart.csv")
df.head(3)
```

```python
x = df.drop("output",axis=1)
y = df["output"]
```

```python
model = DecisionTreeClassifier().fit(x,y)
```

```python
model.score(x,y)
```

**Decision tree models are prone to overfitting. It gave a 100% success rate here too.**

**So we will use train_test_split and see if it is really successful or if it is overfitting.**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=5)
```

```python
model = DecisionTreeClassifier().fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

**Decision Tree Diagram**

```python
dot = export_graphviz(model,feature_names=x.columns)
decision_tree_diagram = graphviz.Source(dot)
decision_tree_diagram
```

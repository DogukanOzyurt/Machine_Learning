```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("insurance.csv")
df.head(3)
```

```python
df.info()
```

```python
df["region"].unique()
```

```python
df = pd.get_dummies(df, columns=["sex","smoker"],drop_first=True)
```

```python
df = pd.get_dummies(df, columns=["region"])
```

```python
df.head(3)
```

```python
x = df.drop("charges",axis=1)
y = df["charges"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=5)
```

```python
model = LinearRegression().fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

# MSE RMSE MAE MAPE

```python
df_error = pd.DataFrame()
df_error["y_test"] = y_test
df_error["y_pred"] = y_pred
df_error.head(5)
```

```python
df_error["error"] = y_test-y_pred
df_error.head(3)
```

**Squared Error**

```python
df_error["squared_error"] = df_error["error"]**2
df_error.head(3)
```

**Absolute Error**

```python
df_error["abs_error"] = np.abs(df_error["error"])
df_error.head(3)
```

**Absolute Percentage Error**

```python
df_error["percent_error"] = np.abs((y_test-y_pred)/y_test)
df_error.head(3)
```

```python
df_error.mean()
```

**Root Mean Square Error**

```python
np.sqrt(df_error["squared_error"].mean())
```

## Error calculation with Sklearn

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

**MSE (Mean Squared Error)**

```python
mean_squared_error(y_test, y_pred)
```

```python
**MAE (Mean Absolute Error)**
```

```python
mean_absolute_error(y_test, y_pred)
```

**MAPE (Mean Absolute Percentage Error)**

```python
mean_absolute_percentage_error(y_test, y_pred)
```

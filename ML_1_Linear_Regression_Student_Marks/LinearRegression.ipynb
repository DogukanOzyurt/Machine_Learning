**y = x + b  -->  y = x1 + x2 + b**

```python
y = df[["Marks"]]
x = df[["number_courses","time_study"]]
```

```python
model = LinearRegression().fit(x,y)
```

```python
model.score(x,y)
```

```python
model.coef_ # x1, x2
```

```python
model.intercept_ # b
```

```python
(np.sum(np.array([4,4]) * model.coef_ ) + model.intercept_)[0]
```

```python
model.predict([[4,4]])[0][0]
```
